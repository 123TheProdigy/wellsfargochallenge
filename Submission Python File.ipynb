```python
#purpose of this function is to clean up the data, filter it out, removing unnecessary features 
def removeOnes(df):
    rev = list(df['trans_desc'])
    for item in range(len(rev)):
        rev[item] = re.sub(" +", " ", rev[item])
        rev[item] = re.sub("111111XXXXXX1111 111111111111111", "", rev[item])
        rev[item] = re.sub("\?MCC=1111 11", "", rev[item])
        rev[item] = re.sub("111-111-1111", " ", rev[item])
        rev[item] = re.sub("111-1111111", "", rev[item])
        rev[item] = re.sub("11111", "", rev[item])
        rev[item] = re.sub("\?MCC=1111", "", rev[item])
        
        
    rev_location = list(df['default_location'])
    rev_location = [str(p) for p in rev_location]
    for item in range(len(rev_location)):
        rev_location[item] = re.sub("111-111-1111", "", rev_location[item])
        rev_location[item] = re.sub("111-1111111", "", rev_location[item])
        rev_location[item] = re.sub("11111", "", rev_location[item])
        
    
        
    for index, row in df.iterrows():
        df.at[index, 'trans_desc'] = rev[index]
        df.at[index, 'default_location'] = rev_location[index]
    return df
```

```python
#prepares inputs for BERT, essentially vectorizing the inputs 
def encoder(allData, length, tokenizer):
    allTokens = []
    allMasks = []
    allSegments = []
    for point in allData:
        point = tokenizer.tokenize(point)
        point = point[:length-2] #making room for CLS and SEP topens 
        pad = ["[CLS]"] + point + ["[SEP]"] #start of sentence and end of sentence
        numPad = length - len(pad) #create vectors of the same size for NN
        
        token = tokenizer.convert_tokens_to_ids(pad) + ([0]*numPad) #create input IDs out of tokens 
        mask = ([1] * len(pad)) + ([0] * numPad) #masking tokens 
        segment = [0] * length #segment tokens 
        
        allTokens.append(token)
        allMasks.append(mask)
        allSegments.append(segment)
    
    return np.array(allTokens), np.array(allMasks), np.array(allSegments)
```

```python
def get_score(X_train, X_test, y_train, y_test):
    X_train = [''.join(i) for i in X_train] #list form
    X_test = [''.join(i) for i in X_test]
    
    train_TEST1 = encoder(X_train, max_length, tokenizer) #run the encoder so inputs can run through BERT
    test_TEST1 = encoder(X_test, max_length, tokenizer)
    
    y_train = tf.keras.utils.to_categorical(y_train, num_classes=11) #set number of categories 
    y_test = tf.keras.utils.to_categorical(y_test, num_classes=11)
    
    early_stopping = EarlyStopping(min_delta = 0.001,
                                   mode = 'max',
                                   monitor = 'accuracy',
                                   patience = 10, restore_best_weights = True) #stop running if model doesn't learn after 10 epochs
    callback = [early_stopping]
    
    num_epochs = 14

    history = model.fit(train_TEST1,y_train,
                        validation_data = (test_TEST1, y_test),
                        epochs = num_epochs,
                        batch_size = 256,
                        verbose = 1,
                        callbacks = callback)
    
    y_test_max = [np.argmax(i) for i in y_test]
    y_pred = model.predict(test_TEST1, verbose = 1)
    y_pred_max = [np.argmax(i) for i in y_pred]
    report = classification_report(y_test_max, y_pred_max) #check recall and precision, forgot to input it here in the final submission but had it running in other files to test
    print(report)
    
    return (model.evaluate(test_TEST1, y_test))
```
