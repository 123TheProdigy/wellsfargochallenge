```python
def removeOnes(df):
    rev = list(df['trans_desc'])
    for item in range(len(rev)):
        rev[item] = re.sub(" +", " ", rev[item])
        rev[item] = re.sub("111111XXXXXX1111 111111111111111", "", rev[item])
        rev[item] = re.sub("\?MCC=1111 11", "", rev[item])
        rev[item] = re.sub("111-111-1111", " ", rev[item])
        rev[item] = re.sub("111-1111111", "", rev[item])
        
    rev_location = list(df['default_location'])
    rev_location = [str(p) for p in rev_location]
    for item in range(len(rev_location)):
        rev_location[item] = re.sub("111-111-1111", "", rev_location[item])
        rev_location[item] = re.sub("111-1111111", "", rev_location[item])
        
    
        
    for index, row in df.iterrows():
        df.at[index, 'trans_desc'] = rev[index]
        df.at[index, 'default_location'] = rev_location[index]
    return df
```

```python
def encoder(allData, tokenizer):
    allTokens = []
    allMasks = []
    allSegments = []
    for point in allData:
        point = tokenizer.tokenize(point)
        point = point[:510]
        pad = ["[CLS]"] + point + ["[SEP]"]
        numPad = 512 - len(pad)
        
        token = tokenizer.convert_tokens_to_ids(pad) + ([0]*numPad)
        mask = ([1] * len(pad)) + ([0] * numPad)
        segment = [0] * 512
        
        allTokens.append(token)
        allMasks.append(mask)
        allSegments.append(segment)
    
    return np.array(allTokens), np.array(allMasks), np.array(allSegments)
```

```python
input_word_ids = tf.keras.Input(shape=(512), dtype=tf.int32, name="input_word_ids")
input_mask = tf.keras.Input(shape=(512), dtype=tf.int32, name="input_mask")
segment_ids = tf.keras.Input(shape=(512), dtype=tf.int32, name="segment_ids")

pooled_output, sequence_output = bert_enc([input_word_ids, input_mask, segment_ids])
clf_output = sequence_output[:, 0, :]
lay = tf.keras.layers.Dropout(0.2)(clf_output)
lay = tf.keras.layers.Dense(32, activation='relu')(lay)
lay = tf.keras.layers.Dropout(0.2)(lay)
out = tf.keras.layers.Dense(11, activation = 'sigmoid')(lay)

model = tf.keras.models.Model(inputs=[input_word_ids, input_mask, segment_ids], outputs=out)
model.compile(tf.keras.optimizers.Adam(), loss='categorical_crossentropy')
```
